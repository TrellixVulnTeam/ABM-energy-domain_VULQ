# Diffusion of Behaviour 

## Jensen 2015: Agent-based assessment framework for behavior-changing feedback devices: Spreading of devices and heating behavior

Model difuziji tehnologije dodaja še difuzijo obnašanja (Prirejeno po medelu Anderson 2014).

- Združevanje dveh modelov:
    - Anderson et. al (2014) → Energy consumption behaviour
        - Behaviour diffusion
            - Previous energy consumption
            - Previous consumption of peers + 
            - Susceptibility (dovzetnost) to subjective norms 
    - Schwarz and Ernst (2009) → Diffusion of sustainable household energy
        - Technology diffusion:
            - Attitude towards the product
            - Subjective Norm
            - Percieved behaviour control

- Za preprostejšo implementacijo enačbe bomo uporabili samo tri agente, obnašanje pa se posodablja samo za prvega v seznamu.

- Obnašanje je definirano z spremembo temperature med 21 in 18 stopinjami, tako kot v članku Jensen 2014.  

```python
#agent = Behaviour, susceptibility, social connection strength 
agent1 = [21, 0.1, 0.3]
agent2 = [18, 0.6, 0.3]
agent3 = [18, 0.6, 0.3]

agentList = [agent1,agent2,agent3] #agente spravimo v seznam

#print(agentList[1:])

plus = sum(agent[0]*agent[2] for agent in agentList[1:]) #Kalkulacija za testiranje 
print(plus)

def calculateBehaviour(time,agent):
    agentBehaviour = agent[0] #Nastavimo začetno obnašanje agenta
    print('Parametri', time, agentBehaviour) 
    for i in range(time): # zanka za simulacijo poteka časa
        agentBehaviour = agentBehaviour + agent[1] * (sum(agent[0]*agent[2] for agent in agentList[1:])/sum(agent[2] for agent in agentList[1:])-agentBehaviour)
        print(agentBehaviour)
```

### 2.1 Implementacija preprostega modela

Za prvo implementacijo smo pripravili preprost model, ki ustvari agente z obnašanjem 18 ali 21. Agenti nato med sabo naključno komunicirajo in ustrezno menjajo obnašanje. 

```python
from copy import deepcopy 
```

Najprej določimo vrednosti, ki jih bomo dodelili agentom. To sta *Behaviour*, ki predstavlja temperaturo, ki jo ima agent nastavljeno v hiši in *Susceptibility*, ki določa koliko je agent dovzeten za spremembo obnašanja. Večja kot je vrednost bolj je dovzeten.

Sledijo funkcije, ki jih bomo kasneje uporabili v simulaciji:
1. **createAgent**: Ustvari agenta v preprostem listu, z dvema zgoraj opisanima parametroma.
2. **createPopulation**: Ustvari populacijo z številom agentov. Obnašanje populacije se izibra naključno med 18 in 21, medtem ko je Susceptibility fiksiran
3. **communication**:V celotni populaciji poišče 2 naključna agenta in jih poveže v par. 
4. **changeBehaviour**: Izračuna nov *Behaviour* za agenta

```python
behaviour = [18.,21.]
susceptibility = [0.3]

#testna agenta
agent1 = [21., 0.3]
agent2 = [18., 0.3]

def createAgent(behaviour, susceptibility):
    return [behaviour, susceptibility]

def createPopulation(agentNum): # ustvari populacijo agento
    ppl = []
    
    for i in range(agentNum):
        ranB = random.randint(0,1)
        susc = 0.3
        agent = createAgent(behaviour[ranB], susceptibility[0])
        ppl.append(agent)
    return ppl

def communication(ppl): #poišče naključne pare v celotni populaciji
    pplSize = len(ppl)-1
    n = random.randint(0, pplSize)
    m = random.randint(0, pplSize)
    while n == m:
        m = random.randint(0, pplSize)
    return ppl[n],ppl[m]

def changeBehaviour(agent1, agent2): #po poenostavljeni formuli spremeni obnašanje
    newBehaviour = agent1[0] + agent2[1]*(agent2[0]-agent1[0]) 
    agent1[0] = deepcopy(newBehaviour)
    #print("New B:", newBehaviour)
```

```python
#Testni blokec
#populacija = createPopulation(10)
#print(populacija)
#communication(populacija)
```

Z zgoraj pripravljenimi funkcijami lahko izvedemo simulacijo in izrišemo histogram s katerim bomo preverili kako se obnašanje razlikuje na začetku in koncu simulacije.

```python
def simulation(agents, time):
    startPpl = createPopulation(agents)
    ppl = deepcopy(startPpl)
    
    for i in range(time):
        pair = communication(ppl)
        changeBehaviour(pair[0],pair[1])
    #enacba brez utezi
    
    
    return startPpl, ppl
```

Poženemo simulacijo in izrišemo graf.

```python
startState, endState = simulation(10,100) #(št agentov, časovni koraki)
```

```python
startBehavior = []
endBehavior = []

for agent in range(len(startState)):
    startBehavior.append(startState[agent][0])
    endBehavior.append(endState[agent][0])
    
plt.hist(startBehavior, color='r')
plt.hist(endBehavior, color='g')
plt.show()
```

### 2.2 Implementacija zahtevnejšim omrežjem
Sledi implementacija v bolj realno okolje. Za to bomo uporabili small world network generiran s pomočjo knjižnice NetworkX. Najprej bomo generirali omrežje in iz njega pridobili podatke, ki jih bomo uporabili za komunikacijo med agenti.

```python
import networkx as nx
import pandas as pd
```

#### 2.2.1 Postavljanje omrežja
Generiraj small-world omrežje: Spodnja funkcija izdela small owrld omrežje po algoritmu Watts-Strogatz. V funkcijo vnesemo število agentov v omrežju. Znotraj metode pa lahko za generiranje omrežja pa lahko spreminjamo še parametra:
- k : Vsak agent bo povezan z k številom drugih agentov
- p : Verjetnost ponovne povezave

[Povezava do metode](https://networkx.github.io/documentation/networkx-1.9/reference/generated/networkx.generators.random_graphs.watts_strogatz_graph.html)

```python
def createNetwork(agentNumber):
    # uporabi algoritem za generiranje omrežja
    network = nx.watts_strogatz_graph(n = agentNumber, k = 2, p = 0.3) 
    # člene postavi na krožnico
    pos = nx.circular_layout(network) 
    # izriši graf
    plt.figure(figsize = (9, 6)) 
    nx.draw_networkx(network,pos)  
    
    return network
```

Funkcijo uporabimo in izrišemo graf:

```python
omrezje = createNetwork(30)
```

#### 2.2.2 Inicializacija agentov 
Smatramo, da je vsak "node" oz. člen v omrežju en agent. Če preberemo podatke znotraj teh agentov, lahko vidimo da so le ti popolnoma prazni:

```python
omrezje.nodes.data()
```

Z naslednjo funkcijo se sprehodimo čez vse agente v omrežju in vsakemu izmed njih dodamo naslednje atribute:
 1. **Behaviour** -> Se naključno izbere med 18 in 21
 2. **Susceptibility** -> Je nastavljen na 0.3
 3. **w** -> Utež družbenih povezav, ki je nastavljena na 0.2 
 4. **neighbours** -> Seznam vseh sosedov posameznega agenta

```python
#inicializacija agentov -> Dodajanje vrednosti
def addAgentValuesToNetwork(network):    
    behaviourList = [18.,21.]
    for node in network.nodes:
        ranB = random.randint(0,1)
        network.nodes[node]['behaviour'] = behaviourList[ranB]
        network.nodes[node]['susceptibility'] = 0.3
        network.nodes[node]['w'] = 0.2
        network.nodes[node]['neighbours'] = list(network.neighbors(node))
    return network
```

Izvedemo funkcijo in ponovno preverimo podatke v agentih:

```python
omrezje = addAgentValuesToNetwork(omrezje)
omrezje.nodes.data()
```

Graf lahko še izrišemo tako, da agentom ki imajo nastavljen Behaviour na 18 dodamo modro barvo, tistim, ki imajo 21 pa rdečo. Koda za rešitev z barvami je dostopna na spodnjem linku:
[Link do rešitve z barvami](https://stackoverflow.com/questions/28910766/python-networkx-set-node-color-automatically-based-on-number-of-attribute-opt)

```python
from itertools import count

groups = set(nx.get_node_attributes(omrezje,'behaviour').values())
mapping = dict(zip(sorted(groups),count()))
nodes = omrezje.nodes()
colors = [mapping[omrezje.node[n]['behaviour']] for n in nodes]

#colors = [i/len(omrezje.nodes) for i in range(len(omrezje.nodes))]
#colormap = nx.get_node_attributes(omrezje, 'behaviour')
pos = nx.circular_layout(omrezje)  
plt.figure(figsize = (9, 6)) 
nx.draw_networkx(omrezje,pos, node_color=colors, cmap = plt.cm.jet)
```

#### 2.2.3. Implementacija enačbe za Behaviour diffusion
V spodnjo funkcijo implementiramo enačbo za iz<račun obnašanja posameznega agenta. Funkcija kot parameter vzame  ID agenta in omrežje v katerem se nahaja. 

Deluje po korakih: 

 1. Ustvari seznam z agentovimi sosedi, ter prazna seznama v katera bomo dodali vsa obnašanja sosedov in vse uteži sosedov
 2. Z zanko se sprehodimo skozi vse agentove sosede in ustrezno napolnimo prej omenjena sesznama. 
 3. Izračunamo novo vrednost **agentBehaviour** po enačbi: ![](https://paper-attachments.dropbox.com/s_A831B714AFE1BEFFE0B96A138A7F786D45DBC26AD537D3FEF7D5EDBD3688FD3C_1592428047496_image.png)
 4. Po izračunu moramo vrednost še posodobiti v slovarju omrežja

```python
def calculateBehaviour(agent, network):
    # naredi seznam agentovih sosedov 
    neighboursList = network.nodes[agent]['neighbours']
    behList = []
    wList = []
    
    #print("Zacetni", network.nodes[agent]['behaviour'])
    
    # pridobi behavior in w vsakega soseda in ga shrani v seznam
    for n in neighboursList:
        behList.append(network.nodes[n]['behaviour'])
        wList.append(network.nodes[n]['w'])
    #print(behList)

    # izracunaj behaviour po enacbi
    agentBehaviour = network.nodes[agent]['behaviour'] + network.nodes[agent]['susceptibility'] * (
                sum(behList[n] * wList[n] for n in range(len(behList))) / sum(wList) - network.nodes[agent][
            'behaviour'])

    print("Izracunan", agentBehaviour)

    # updejtaj agentov behaviour v omrežju
    network.nodes[agent]['behaviour'] = agentBehaviour

    return network.nodes[agent]['behaviour']
```

```python
# Testni blok
# print(behList)
calculateBehaviour(6, omrezje)
```

#### 2.2.4. Simulacija
Spodnja funkcija se sprehodi skozi vse agente v omrežju in računa njihovo obnašanje v določenem številu časovnih korakov.  

```python
def communicatePeers(network,timesteps):
    t = 0
    while t < timesteps:
        for agents in network:
            print("Agent:", agents)
            calculateBehaviour(agents,network)
        t += 1
```

Izvedemo funkcijo in izrišemo graf s pomočjo barv iz prejšnega izrisa, tako da dobimo gradient barv in lahko vidimo, kako se agentom skozi čas izvajanja obnašanje zleze proti sredini med 18 in 21. 

```python
communicatePeers(omrezje,10)
```
